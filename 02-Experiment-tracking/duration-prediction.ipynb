```python
# Lê um arquivo Parquet contendo os dados de corridas de táxi verdes do mês de janeiro de 2021
df = pd.read_parquet('/workspaces/mlops-techniques/01-Environment/data/green_tripdata_2021-01.parquet')

# Calcula a duração de cada corrida subtraindo o horário de início (pickup) do horário de término (dropoff)
df['duration'] = df.lpep_dropoff_datetime - df.lpep_pickup_datetime

# Converte a duração de timedelta para minutos
df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)

# Filtra o DataFrame para manter apenas as corridas com duração entre 1 e 60 minutos
df = df[(df.duration >= 1) & (df.duration <= 60)]

# Define as colunas categóricas (IDs de localização de origem e destino)
categorical = ['PULocationID', 'DOLocationID']

# Define as colunas numéricas (distância da viagem)
numerical = ['trip_distance']

# Converte as colunas categóricas para o tipo string
df[categorical] = df[categorical].astype(str)
```

```python
# Converte as colunas categóricas e numéricas do DataFrame para uma lista de dicionários
# Cada dicionário representa uma observação (linha) no formato {coluna: valor}
train_dicts = df[categorical + numerical].to_dict(orient='records')

# Cria um vetorizador para transformar os dicionários em matrizes numéricas
dv = DictVectorizer()

# Aplica o vetorizador ao conjunto de dados, ajustando e transformando as colunas categóricas e numéricas em um formato apropriado para modelos
X_train = dv.fit_transform(train_dicts)

# Define a variável de destino (target) como sendo a coluna 'duration'
target = 'duration'

# Extrai os valores da variável de destino como um array NumPy
y_train = df[target].values

# Cria uma instância do modelo de regressão linear
lr = LinearRegression()

# Ajusta o modelo usando os dados de entrada (X_train) e o target (y_train)
lr.fit(X_train, y_train)

# Realiza previsões no conjunto de treino
y_pred = lr.predict(X_train)

# Calcula o erro quadrático médio raiz (RMSE) para avaliar a qualidade das previsões
root_mean_squared_error(y_train, y_pred)
```

```python
# Cria um gráfico de distribuição (densidade) para as previsões feitas pelo modelo (y_pred)
# label='prediction' define o rótulo que será exibido na legenda
sns.distplot(y_pred, label='prediction')

# Cria um gráfico de distribuição (densidade) para os valores reais do conjunto de treino (y_train)
# label='actual' define o rótulo que será exibido na legenda
sns.distplot(y_train, label='actual')

# Adiciona uma legenda ao gráfico para identificar as distribuições
plt.legend()
```

```python
def read_dataframe(filename):
    # Verifica se o arquivo é um CSV
    if filename.endswith('.csv'):
        # Lê o arquivo CSV em um DataFrame
        df = pd.read_csv(filename)
        
        # Converte as colunas de data e hora para o formato datetime
        df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
        df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
    
    # Verifica se o arquivo é um Parquet
    elif filename.endswith('.parquet'):
        # Lê o arquivo Parquet em um DataFrame
        df = pd.read_parquet(filename)

    # Calcula a duração da viagem como a diferença entre as datas de drop-off e pickup
    df['duration'] = df.lpep_dropoff_datetime - df.lpep_pickup_datetime

    # Converte a duração para minutos
    df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)

    # Filtra viagens com duração entre 1 e 60 minutos
    df = df[(df.duration >= 1) & (df.duration <= 60)]

    # Define as colunas categóricas
    categorical = ['PULocationID', 'DOLocationID']

    # Converte as colunas categóricas para o tipo string
    df[categorical] = df[categorical].astype(str)
    
    # Retorna o DataFrame processado
    return df
```

```python
df_train = read_dataframe('/workspaces/mlops-techniques/01-Environment/data/green_tripdata_2021-01.parquet')
df_val = read_dataframe('/workspaces/mlops-techniques/01-Environment/data/green_tripdata_2021-02.parquet')
```

```python
len(df_train), len(df_val)
```

```python
# Cria uma nova coluna no DataFrame de treino chamada 'PU_DO'
# Combina as colunas 'PULocationID' e 'DOLocationID' em um único identificador,
# separadas por um underline ('_').
df_train['PU_DO'] = df_train['PULocationID'] + '_' + df_train['DOLocationID']

# Faz o mesmo processo acima para o DataFrame de validação.
df_val['PU_DO'] = df_val['PULocationID'] + '_' + df_val['DOLocationID']
```

```python
# Define as colunas categóricas e numéricas
# Substitui as colunas originais 'PULocationID' e 'DOLocationID' pela nova coluna combinada 'PU_DO'.
categorical = ['PU_DO']  # Coluna categórica resultante da combinação de 'PULocationID' e 'DOLocationID'.
numerical = ['trip_distance']  # Coluna numérica que representa a distância da viagem.

# Instancia o DictVectorizer, que será usado para transformar os dados categóricos e numéricos em uma matriz numérica.
dv = DictVectorizer()

# Converte as colunas categóricas e numéricas do conjunto de treino em uma lista de dicionários.
# Cada linha do DataFrame se torna um dicionário com os nomes das colunas como chaves e seus valores correspondentes.
train_dicts = df_train[categorical + numerical].to_dict(orient='records')

# Ajusta o DictVectorizer com os dados de treino e transforma esses dicionários em uma matriz esparsa.
# As colunas categóricas são transformadas em variáveis dummy (codificação one-hot).
X_train = dv.fit_transform(train_dicts)

# Faz o mesmo processo para o conjunto de validação, mas usa o modelo já ajustado (fit) no conjunto de treino.
val_dicts = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dicts)
```

```python
target = 'duration'
y_train = df_train[target].values
y_val = df_val[target].values
```

```python
# Instancia o modelo de Regressão Linear
lr = LinearRegression()

# Ajusta o modelo de regressão linear aos dados de treino
# O modelo aprende a relação entre as variáveis independentes (X_train) e a variável dependente (y_train)
lr.fit(X_train, y_train)

# Faz a previsão da variável dependente para o conjunto de validação (X_val)
y_pred = lr.predict(X_val)

# Calcula o erro quadrático médio (RMSE) entre os valores reais (y_val) e as previsões (y_pred)
# RMSE é uma métrica de avaliação de modelos de regressão, que mede a diferença média entre os valores previstos e os valores reais
root_mean_squared_error(y_val, y_pred)
```

```python
import xgboost as xgb

from hyperopt import fmin, tpe, hp, STATUS_OK, Trials
from hyperopt.pyll import scope
```

```python
train = xgb.DMatrix(X_train, label=y_train)
valid = xgb.DMatrix(X_val, label=y_val)
```

def objective(params):  
    # Define uma função `objective` que será usada como objetivo para um processo de otimização de hiperparâmetros.
    # O argumento `params` contém os hiperparâmetros que serão ajustados.

    with mlflow.start_run():  
        # Inicia um novo experimento no MLflow para registrar os detalhes da execução.

        mlflow.set_tag("model", "xgboost")  
        # Define uma tag no MLflow indicando que o modelo utilizado é o "xgboost".

        mlflow.log_params(params)  
        # Registra os hiperparâmetros fornecidos (`params`) como parâmetros no MLflow.

        booster = xgb.train(  
            params=params,  
            dtrain=train,  
            num_boost_round=1000,  
            evals=[(valid, 'validation')],  
            early_stopping_rounds=50  
        )  
        # Treina um modelo XGBoost:
        # - `params`: os hiperparâmetros fornecidos.
        # - `dtrain`: o conjunto de treinamento.
        # - `num_boost_round`: número máximo de iterações (árvores) que o modelo será treinado.
        # - `evals`: conjunto de validação para monitorar o desempenho durante o treinamento.
        # - `early_stopping_rounds`: para interromper o treinamento cedo se o desempenho não melhorar após 50 iterações consecutivas.

        y_pred = booster.predict(valid)  
        # Faz previsões no conjunto de validação usando o modelo treinado.

        rmse = mean_squared_error(y_val, y_pred, squared=False)  
        # Calcula o RMSE (Root Mean Squared Error) das previsões em relação aos valores reais de validação.
        # O parâmetro `squared=False` retorna a raiz quadrada do erro quadrático médio.

        mlflow.log_metric("rmse", rmse)  
        # Registra o RMSE como uma métrica no MLflow.

    return {'loss': rmse, 'status': STATUS_OK}  
    # Retorna um dicionário com:
    # - `loss`: o valor do RMSE (usado pelo otimizador para avaliar o desempenho do modelo).
    # - `status`: indica que a execução foi bem-sucedida (geralmente usado em frameworks como Hyperopt).


```python
def objective(params):
    with mlflow.start_run():
        mlflow.set_tag("model", "xgboost")
        mlflow.log_params(params)
        booster = xgb.train(
            params=params,
            dtrain=train,
            num_boost_round=1000,
            evals=[(valid, 'validation')],
            early_stopping_rounds=50
        )
        y_pred = booster.predict(valid)
        rmse = np.sqrt(mean_squared_error(y_val, y_pred))  
        mlflow.log_metric("rmse", rmse)

    return {'loss': rmse, 'status': STATUS_OK}
```

```python
#vai demorar 2h - testei 5
search_space = {
    'max_depth': scope.int(hp.quniform('max_depth', 4, 100, 1)),
    'learning_rate': hp.loguniform('learning_rate', -3, 0),
    'reg_alpha': hp.loguniform('reg_alpha', -5, -1),
    'reg_lambda': hp.loguniform('reg_lambda', -6, -1),
    'min_child_weight': hp.loguniform('min_child_weight', -1, 3),
    'objective': 'reg:linear',
    'seed': 42
}

best_result = fmin(
    fn=objective,
    space=search_space,
    algo=tpe.suggest,
    max_evals=50,
    trials=Trials()
)
```
